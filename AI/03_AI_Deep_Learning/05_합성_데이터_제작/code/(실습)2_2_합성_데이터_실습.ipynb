3. pythonì€ ê¸°ë³¸ì ìœ¼ë¡œ ë™ê¸°ì ìœ¼ë¡œ ë™ì‘í•˜ê¸° ë•Œë¬¸ì— ì´ëŠ” ë§¤ìš° ë¹„íš¨ìœ¨ì ì…ë‹ˆë‹¤.
4. ë¹„ë™ê¸° API í˜¸ì¶œì„ í•˜ê²Œ ë˜ë©´ ì½”ë“œìƒì—ì„œ í•œë²ˆì— ì—¬ëŸ¬ APIë¥¼ í˜¸ì¶œí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ì´ëŸ¬í•œ ì´ìœ ë¡œ ë¹„ë™ê¸° í†µì‹ ì„ ì‚¬ìš©í•˜ëŠ” ê²ƒì´ ë§¤ìš° ì¤‘ìš”í•©ë‹ˆë‹¤. ê·¸ëŸ¬ë©´ ì•„ë˜ ì½”ë“œë¡œ ë¹„ë™ê¸° í†µì‹ ì„ êµ¬í˜„í•´ë³´ê² ìŠµë‹ˆë‹¤.

````python
import httpx

async def call_chat_completion(url: str, headers: dict, payload: dict):
    async with httpx.AsyncClient(timeout=30.0) as client:
        response = await client.post(url, headers=headers, json=payload)
        response.raise_for_status()
        data = response.json()

        return data["choices"][0]["message"]["content"]
````

ìœ„ì—ì„œ ì²˜ëŸ¼ `async with httpx.AsyncClient() as client:` ì„ ì‚¬ìš©í•˜ì—¬ REST APIë°©ì‹ìœ¼ë¡œ ë¹„ë™ê¸°ì ìœ¼ë¡œ í˜¸ì¶œì„ í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ê·¸ëŸ¬ë©´ ì‹¤ì œë¡œ íŒŒë¼ë¯¸í„° ê°’ë“¤ì„ ë„£ê³  ê°„ë‹¨í•œ í˜¸ì¶œì„ ì§„í–‰í•´ë³´ê² ìŠµë‹ˆë‹¤.

````python
import asyncio
# https://console.upstage.ai/api/chat -> cURL
async def main():
    url = "https://api.upstage.ai/v1/chat/completions"
    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {UPSTAGE_API_KEY}"
    }
    prompts = [
        "Tell me a joke about cats",
        "What is the capital of France?",
        "Summarize the plot of Inception in 2 sentences."
    ]

    tasks = []
    for prompt in prompts:
        payload = {
            "model": "solar-pro2",
            "messages": [
                {
                    "role": "user",
                    "content": prompt
                }
            ],
            "stream": False
        }
        tasks.append(call_chat_completion(url, headers, payload))

    results = await asyncio.gather(*tasks)

    for i, res in enumerate(results, 1):
        print(f"--- Response {i} ---")
        print(res)
        print()

# ì‹¤í–‰
await main()
````

3ê°œì˜ ì‘ë‹µì„ í•œë²ˆì— ë°›ì„ ìˆ˜ ìˆì—ˆìŠµë‹ˆë‹¤.

ìœ„ì˜ ì½”ë“œëŠ” ì•„ë˜ python ì½”ë“œì™€ ë™ì¼í•œ ê²°ê³¼ë¥¼ ì–»ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
from openai import AsyncOpenAI
# https://console.upstage.ai/api/chat -> python
# OpenAI ë¼ì´ë¸ŒëŸ¬ë¦¬(í´ë¼ì´ì–¸íŠ¸)ë¥¼ ì¨ì„œ Upstageì˜ ëª¨ë¸ì— ìš”ì²­í•˜ëŠ” ë°©ì‹
# ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ í†µí•´ì„œ ì—”ë“œí¬ì¸íŠ¸ë§Œ ë°”ê¿”ì¤˜ì„œ ì‚¬ìš©í•  ìˆ˜ ìˆê²Œ ë§Œë“  ê²ƒ

client = AsyncOpenAI(api_key=UPSTAGE_API_KEY, base_url="https://api.upstage.ai/v1")

async def chat_completion(prompt: str, model: str = "solar-pro2") -> str:
    """
    ë¹„(é)ìŠ¤íŠ¸ë¦¬ë° í˜¸ì¶œ ë²„ì „. í•œ ë²ˆì— ì „ì²´ ì‘ë‹µì„ ë°›ì•„ì˜µë‹ˆë‹¤.
    """
    resp = await client.chat.completions.create(
        model=model,
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        stream=False,
    )
    return resp.choices[0].message.content

async def main():
    prompts = [
        "Tell me a joke about cats",
        "What is the capital of France?",
        "Summarize the plot of Inception in 2 sentences."
    ]

    # ë™ì‹œì— ìŠ¤íŠ¸ë¦¬ë° í˜¸ì¶œ
    tasks = [chat_completion(p) for p in prompts]
    results = await asyncio.gather(*tasks)

    # ê²°ê³¼ ì •ë¦¬ ì¶œë ¥
    for i, res in enumerate(results, 1):
        print(f"--- Response {i} (collected) ---")
        print(res)
        print()

# Colab/Jupyterì—ì„œëŠ” ìµœìƒë‹¨ ì…€ì—ì„œ ë°”ë¡œ ì‹¤í–‰ ê°€ëŠ¥
await main()
````

````python
from openai import OpenAI
import json

client = OpenAI(
    api_key=UPSTAGE_API_KEY,
    base_url="https://api.upstage.ai/v1"
)

user_prompt = """ê°€ìƒì˜ ë°ì´í„°ë¥¼ ë§Œë“¤ë ¤ê³  í•©ë‹ˆë‹¤.

ì•„ë˜ output_formatì— ë”°ë¼ JSON ë°ì´í„°ë¥¼ 1ê°œë§Œ ë°˜í™˜í•´ì£¼ì„¸ìš”.
{output_format}
"""

# TODO: output_formatì„ ì •ì˜í•´ì£¼ì„¸ìš”.
output_format ="""<output_format>
```json
{
    "name": [ì´ë¦„],
    "age": [ë‚˜ì´]
    "is_student": [true ë˜ëŠ” false]
}
```
</output_format>"""

messages=[
    {
        "role": "user",
        "content": user_prompt.format(output_format=output_format)
    }
]

response = client.chat.completions.create(
    model="solar-pro2",
    messages=messages,
)

def json_parsing(output_text:str) -> dict:
    output_text = output_text[output_text.index("```json") + len("```json"):].strip()
    output_text = output_text[:output_text.index("```")].strip()
    return json.loads(output_text)


output = response.choices[0].message.content
print("í˜„ì¬ ì‘ë‹µì˜ í˜•ì‹: ", type(output))
print(output)
structured_dictionary = json_parsing(output)
print("êµ¬ì¡°í™”ëœ ì‘ë‹µì˜ í˜•ì‹: ", type(structured_dictionary))
print(structured_dictionary)
````
